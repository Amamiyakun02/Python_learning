```python
import matplotlib.pyplot as plt
import numpy as np

data = {'a' : np.arange(40),
        'c' : np.random.randint(0,40,40),
        'd' : np.random.randn(40)}
data['b'] = data['a'] + 5 * np.random.randn(40)
data['d'] = np.abs(data['d']) * 80

plt.scatter('a','b', c = 'c', s='d',data=data)
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.show()
```
